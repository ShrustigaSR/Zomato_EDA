```python
cuisine_df = final_df['Cuisines'].str.split(', ', expand=True).stack().reset_index(level=1, drop=True).rename('Cuisine')
```

```python
cuisine_df
```

```python
top_cuisines = cuisine_df.value_counts().head(10)
```

Top 10 most popular cuisines based on frequency

```python
top_cuisines
```

```python
sns.barplot(x = top_cuisines.index,y = top_cuisines.values)
plt.xlabel("Cuisines")
plt.ylabel("Count")
plt.title("Top 10 cuisines based on number of restaurants listed")
plt.show()
```

```python
temp_df = final_df.merge(cuisine_df, left_index=True, right_index=True)
```

```python
temp_df
```

```python
temp_df.drop(columns = "Cuisines", inplace = True)
```
